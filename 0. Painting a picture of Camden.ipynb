```python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from requests import get
from urllib.parse import urlparse
```

```python
GLA = gpd.read_file("https://data.london.gov.uk/download/gla-high-street-boundaries/166e8e42-c1ea-42ec-bd1d-99dc29d04fa8/GLA_High_Street_boundaries.gpkg")
```

```python
GLA.plot()
```

```python
boros = gpd.read_file('data/geo/Boroughs.gpkg')
```

```python
Camden = boros[boros["NAME"]=="Camden"]
```

```python
CamHS = gpd.clip(GLA, Camden)
```

```python
CamHS.to_file('OSM Camden/Camden HighStreets.shp')
```

### Hampstead Example

```python
hamp = gpd.read_file('High Street Boundaries/Hampstead.shp')
```

```python
hamp.plot()
```

```python
hampB = gpd.clip(buildings, hamp)
```

```python
hampB
```

```python
hampP = gpd.clip(pois, hamp)
```

```python
hampR = gpd.clip(religious, hamp)
```

```python
roadsHamp = gpd.clip(roads, hamp)
```

```python
hampB.plot()
hampP.plot()
hampR.plot()
```

```python
HampPoly = gpd.overlay(hampB, hampP, how="union")
```

```python
HampPoly = gpd.overlay(HampPoly, hampR, how="union")
```

```python
HampPoly.plot()
```

```python
Hpoints = gpd.read_file('High Street Points (WHYTHAWK)/Hampstead.shp')
```

```python
Hpoints  = Hpoints[(Hpoints['period'] == '2021-03')]
```

```python
Hpoints
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,9))

hamp.plot(edgecolor="black", facecolor='white', ax=ax)
HampPoly.plot(edgecolor="red", facecolor='red', alpha =0.3, ax=ax)
Hpoints.plot(color= "black", ax=ax)
```

```python
Hampstead = gpd.sjoin(HampPoly, Hpoints, how="left", op="intersects")
```

```python
Hampstead = Hampstead['geometry'].drop_duplicates()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,9))

hamp.plot(edgecolor="black", facecolor='white', ax=ax)
#hampB.plot(edgecolor="red", facecolor='red', alpha =0.3, ax=ax)
Hpoints.plot(color= "black", marker = "s", ax=ax)
Hampstead.plot(edgecolor="red", facecolor='red', alpha =0.3, ax=ax)
roadsHamp.plot(color= "black", ax=ax)
```
